```python
import os
folder_name='Bar_Chart'
os.makedirs(folder_name,exist_ok=True)

director=os.getcwd()
new_dir=os.path.join(director,folder_name)

for i in cat[1:]:
    keys=visa_df[i].value_counts().keys()
    value=visa_df[i].value_counts().values
    plt.figure(figsize=(10,4))
    plt.bar(keys,value)
    plt.xlabel('Classes')
    plt.ylabel('Frequency')
    file_name=f'{i}_bar_chart.jpg'
    path=os.path.join(new_dir,file_name)

    plt.savefig(path)
    plt.close()
```

```python
import os
folder_name='pie_Chart'
os.makedirs(folder_name,exist_ok=True)

director=os.getcwd()
new_dir=os.path.join(director,folder_name)

for i in cat[1:]:
    keys=visa_df[i].value_counts(normalize=True)
    plt.figure(figsize=(10,4))
    plt.pie(keys)
    plt.xlabel('Classes')
    plt.ylabel('Frequency')
    file_name=f'{i}_pie_chart.jpg'
    path=os.path.join(new_dir,file_name)

    plt.savefig(path)
    plt.close()
```

```python
import os
folder_name='pie_Chart'
os.makedirs(folder_name,exist_ok=True)

director=os.getcwd()
new_dir=os.path.join(director,folder_name)

for i,j in enumerate(cat[1:]):
    keys=visa_df[i].value_counts(normalize=True)
    plt.figure(figsize=(10,4))
    lables=list(visa_df[j].dtypes.keys())
    plt.pie(keys,lables=lables)
    plt.xlabel('Classes')
    plt.ylabel('Frequency')
    file_name=f'{i}_pie_chart.jpg'
    path=os.path.join(new_dir,file_name)

    plt.savefig(path)
    plt.close()
```
